[**Apache Spark**](https://spark.apache.org/) is an *open-source* unified analytics engine for large-scale data processing. Spark provides an interface for programming clusters with implicit data parallelism and fault tolerance. While, [**PySpark**](https://spark.apache.org/docs/latest/api/python/#) is the Python API for Apache Spark. As on today (20.04.2023), the latest version available is `3.4.0`. The next steps provide installation guide:

## Dependencies

Spark requires Java 8 or above as mentioned in the [dependency page](https://spark.apache.org/docs/latest/api/python/getting_started/install.html#dependencies). Download and install Java 8 (JDK 20 - is the current version) from [**Oracle Java Downloads**](https://www.oracle.com/in/java/technologies/downloads/), and set the following path environments:

```powershell
set JAVA_HOME = "C:\Program Files\Java\jdk-20"

# PATH Variables
set PATH = %PATH%;"C:\Program Files\Java\jdk-20\bin" # add bin directory
set PATH = %PATH%;"C:\Program Files\Common Files\Oracle\Java\javapath" # javapath
```

## `PySpark` Installation

Install `pyspark` locally using `pip install pyspark`, however it is recommended to **create a new environment** and `jupyter kernel` for the same. More information is available [here](https://spark.apache.org/docs/latest/api/python/getting_started/install.html). To check if the setup is succesfull, run:

```python
try:
    import pyspark
    from pyspark import SparkContext, SparkConf
    from pyspark.sql import SparkSession, SQLContext
except ImportError as e:
    raise ImportError('PySpark is not Configured')
    
print(f"PySpark Version : {pyspark.__version__}")

# define a spark builder object
# https://jovian.com/dpramanik/handling-out-of-memory-exception-in-pyspark
spark = SparkSession.builder.appName("Hello World").getOrCreate()
```

````python
try:
    import pyspark
    from pyspark import SparkContext, SparkConf
    from pyspark.sql import SparkSession, SQLContext
except ImportError as e:
    raise ImportError('PySpark is not Configured')

print(f"PySpark Version : {pyspark.__version__}")
````

````python
spark = SparkSession.builder.appName("Hello World").getOrCreate()
# spark # UI is available at: http://localhost:4040/jobs/
````

## Loading Datasets

Python `sklearn` module provides sample datasets. For this demonstration purpose, *diabetes* dataset is selected and the `pd.DataFrame` is converted to an `pyspark.sql.dataframe.DataFrame` object as below. A CSV file can also be directly read using `pyspark` using `spark.read.csv(*args, **kwargs)` options like:

```python
# reading using `.option()`
df = spark.read.option("header", "true").csv("/path/to/file.csv")

# or can be read using `.csv(**kwargs)` like
df = spark.read.csv("/path/to/file.csv", header = True, inferSchema = True)

# check the contents of `df` using `df.show()` or
df.head()
```

````python
data = spark.read.csv("diabetes.csv", header = True, inferSchema = True)
````

````python
# print the data schema (equivalent to pd.info())
data.printSchema()
````

````python
data.show() # similar to pd.head()
````
