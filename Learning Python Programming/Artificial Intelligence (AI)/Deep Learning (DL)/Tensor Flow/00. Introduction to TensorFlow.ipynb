## Introduction to _tensors_

```python
# constant/scalar
scalar = tf.constant(7)
scalar, scalar.ndim
```

```python
# constant/vector
vector = tf.constant([7, 7])
vector, vector.ndim
```

```python
# variables
var = tf.Variable(7, name = "variable", dtype = tf.float64)
var
```

```python
# random tensors
matrix = tf.random.normal(shape = (10000, 10000, 500), dtype = tf.float64)
matrix, matrix.ndim
```
