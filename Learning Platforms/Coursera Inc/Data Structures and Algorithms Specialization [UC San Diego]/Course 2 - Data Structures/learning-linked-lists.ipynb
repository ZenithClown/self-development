```python
import random
```

```python
from linkedlists import SinglyLinkedList
```

```python
random.seed(7) # set seed for code duplication
llist = SinglyLinkedList()

print(f"Create a Blank Linked List: {llist}")
for _ in range(5):
    llist.append(random.randint(0, 100))
    
print(f"Final Linked List: {llist}")
```

```python
# prepend an element to the linked list
llist.prepend(10) # let add `10` to the beginning
print(f"Final Linked List: {llist}")
```

```python
# find an element from linked list
for element in [75, 41]: # 75 not in llist, however 41 is present
    print(f"{element} present in Linked List." if llist.find(element) else f"{element} is absent.")
```
